The MELD score ranges from 6 to 40 and is derived based on the probability of survival at 3 months for patients with ESLD. Data in ESLD is usually sparse and often aggregated into Stages. We assume that there are 18 stages based on the ESLD: Stage 1, Stage2, ..., Stage 18. The time step is 1 year and the actions in Stages 1 through 18 are "wait" (denoted by 0) and "transplant" (denoted by 1). 

We assume that the Markov property holds. There are two additional states of the Markov Decision Process: "Posttransplant Life" (denoted by 19) and "Death" (which is denoted by 20 and combines so caled "Pretransplant Death" and "Posttransplant Death"). The only action availible in state "Posttransplant Life" is "wait" and "Death" is the terminal state with no actions. Assume that the length of an episode is T=50, unless it terminates earlier due to the transition to the absorbing state "Death."

We do not know the transition probabilities, but if a patient selects "wait," the possible transitions are   
1) Stage 1->Stage 1, Stage 1->Stage 2, Stage 1->Death  
2) For k in {2,3,4,...17}, Stage k->Stage (k-1), Stage k->Stage k, Stage k->Stage (k+1), Stage k->Death    
3) Stage 18->Stage 17, Stage 18->Stage 18, Stage 18->Death    

If a patient selects "transplant" at Stage k, k=1,2,...,18, the only possible transition is  
4) Stage k->"Posttransplant Life"

Finally, there are two more possible transitions"  
5) "Posttransplant Life"->"Posttransplant Life" and "Posttransplant Life"->"Death"  


The patient gets reward 1 in all states "Stage k" (k=1,2,...,18) and reward 0.2 in the "Posttransplant Life" state - assume that the patient gets these rewards on "exit" from the states, i.e. after we observe the corresponding stage. We assume the discounting parameter $\gamma=0.97$, one of the most common discounting rate used in medical decision making (Gold MR, Siegel JE, Russell LB, Weinstein MC. ### (b) Quadratic approximation

(b) Using the train data, please approximate the state-values via the semi-gradient 1-step TD algorithm (with experience replay) for estimating $v_\pi$ with

$$v_\pi(s)\approx\hat{v}(s,{\bf w}) = \sum_{k=1}^{18} \left(w_1+w_2\, \frac{k-9}{5}+w_3\, \left(\frac{k-9}{5}\right)^2\right)\cdot \mathbb{1}_{(s=\text{"Stage }k\text{"})}+w_4\cdot \mathbb{1}_{(s=\text{"Posttransplant Life"})}.$$

Experiment with $\alpha$. Make sure you achieved the minimum. Here, you may want to look at how the weights evolve over time.

```python
def timeit(method):
    def timed(*args, **kw):
        ts = time.time()
        result = method(*args, **kw)
        te = time.time()
        if 'log_time' in kw:
            name = kw.get('log_name', method.__name__.upper())
            kw['log_time'][name] = int((te - ts) * 1000)
        else:
            print ('%r  %2.2f ms' % \
                  (method.__name__, (te - ts) * 1000))
        return result
    return timed
```

```python
def v_hat(s, w):
    k = (s-9)/5
    return (w[0] + w[1]*k + w[2]*k**2)*(s<=18) + w[3]*(s==19)
```

```python
w = np.ones(4)
for s in range(1,21):
    print("state:", s, ", v_hat:",v_hat(s,w))
```

```python
def dv_hat(s):
    k = (s-9)/5
    dv = [1, k, k**2,0]*int(s<=18) + [0, 0, 0, 1]*int(s==19) + [0, 0, 0, 0]*int(s==20)
    return np.array(dv)

for s in range(1,21):
    print("state:", s, ", gradient:",dv_hat(s))
```

```python
def get_episode(df, k):
    """
    Return States, Actions, Rewards from episode in a dataframe
    
    Args:
        df: dataframe
        k: int: kth episode  
        
    Returns:
        list: States, Actions, Rewards
    """
    states = df.values[k,0:-1:2]
    actions = df.values[k,1::2]
    rewards = (states <= 18)*1 + (states == 19)*0.2
    return [states, actions, rewards]
```

```python
@timeit
def semi_gradient_td(Î± = lambda a: 0.1, Î³ = 0.97, df=train, v_hat=v_hat, dv_hat=dv_hat, n_weights = 4, w_init=None, num_epochs = 1000000, seed=1234):
    """
    Semi gradient TD algorithm with eligibility traces
    
    Args:
        Î± (func)         : Alpha function that is evaluated, allows for fixed or flexible weight decays
        Î³ (float)        : discount factor between 0 and 1
        df (dataframe)   : pandas dataframe containing episodes
        v_hat (function) : function to compute v_hat(s, w)
        dv_hat (function): function that computes derivative of v_hat
        n_weights (int)  : specify the number of weights required - 4 for gradient methods, 20 for tabular methods
        w_init (Bool or numpy array): initial weights
        num_epochs (int) : Number of bootstrap samples to take
        seed (int)       : random seed to use for reproducibility
        
    Returns:
        v_hat (list) Estimates of V_hat
        w_replay (numpy array) Final weights
        w_history (numpy array) History of weights
    """
    np.random.seed(seed)
    N = 0
    w_replay = np.zeros(n_weights)
    
    if not w_init is None:
        w_replay = w_init
    
    w_history = w_replay.reshape((n_weights,1))

    for _ in range(num_epochs):
        N += 1
        # sample the episode
        k = np.random.randint(0, df.shape[0])
        episode = get_episode(df, k)
        # sample the state
        t = np.random.randint(low=0, high=49)
        S = episode[0][t]
        S1 = episode[0][t+1]
        R = episode[2][t]
        z = dv_hat(S)
        Î´ = R + Î³ * v_hat(S1, w_replay) - v_hat(S, w_replay)
        w_replay += Î±(N)*Î´*z
        
        if (int(N/1000)*1000==N):
            w_history = np.append(w_history, w_replay.reshape((n_weights,1)), axis=1)
    return [v_hat(k,w_replay) for k in range(1,21)], w_replay, w_history
```

```python
def plot_history(w_history, Î±, title="Semi gradient TD: Dependence of estimated $v_\pi$", ylabel="$v_\pi$", legend=True):
    plt.figure(figsize=(16,6))
    for i in range(w_history.shape[0]):
        plt.plot(range(1,w_history.shape[1]*1000,1000),w_history[i], linestyle='--', marker=None, markersize=4,label=f'$w_{i+1}$')
    
    plt.xlabel("time step $t$",fontsize=16)
    plt.ylabel(ylabel,fontsize=16)
    plt.title(f'{title} on time step $t$ for Î±={Î±}',fontsize=16)
    if legend:
        plt.legend(loc = 'upper left',fontsize=14)
    plt.grid(True)
```

```python
td_1 = semi_gradient_td(Î±=lambda N: 1/N**.5)
```

```python
plot_history(td_1[2], "1/$\sqrt{N}$")
```

```python
td_2 = semi_gradient_td(Î±=lambda N: 1/N**.6)
plot_history(td_2[2], "1/$N^{0.6}$")
```

```python
td_2[0]
```

Here we used $\alpha$ = $\frac{1}{\sqrt{N}}$ to double check our bias, and found that $\alpha$ = $\frac{1}{N^{0.6}}$ gave us the most stable convergence.

### (c) 3rd degree polynomial approximation

(c) Using the train data, please approximate the state-values via the semi-gradient 1-step TD algorithm (with experience replay) for estimating $v_\pi$ with

$$v_\pi(s)\approx\hat{v}(s,{\bf w}) = \sum_{k=1}^{18} \left(w_1+w_2\, \frac{k-9}{5}+w_3\, \left(\frac{k-9}{5}\right)^2+w_4\, \left(\frac{k-9}{5}\right)^3\right)\cdot \mathbb{1}_{(s=\text{"Stage }k\text{"})}+w_5\cdot \mathbb{1}_{(s=\text{"Posttransplant Life"})}.$$

Again, experiment with $\alpha$ (it could be different form (b)); make sure you achieved the minimum.  

```python
def v_hat_3d(s, w):
    k = (s-9)/5
    return (w[0] + w[1]*k + w[2]*k**2 + w[3]*k**3)*(s<=18) + w[4]*(s==19)

w = np.ones(5)
for s in range(1,21):
    print("state:", s, ", v_hat:",v_hat_3d(s,w))
```

```python
def dv_hat_3d(s):
    k = (s-9)/5
    dv = [1, k, k**2,k**3,0]*int(s<=18) + [0, 0, 0, 0, 1]*int(s==19) + [0, 0, 0, 0, 0]*int(s==20)
    return np.array(dv)

for s in range(1,21):
    print("state:", s, ", gradient:",dv_hat_3d(s))
```

```python
td_3 = semi_gradient_td(Î±=lambda N: 1/math.sqrt(N), n_weights=5,v_hat=v_hat_3d,dv_hat=dv_hat_3d)
```

### (d) Validation plot

(d) Using test data, compute observed returns (cumulative discounted rewards) $G_0$ obtained under policy $\pi$. These returns will be used for validation. It is important that the results for the state-value function obtained in (b) and (c) also correspond to $T=50$ (not $T=\infty$). Please notice that we will have 1,600 retunrs only (in fact, one could also consider $G_1$, $G_2$, etc.). In one graph, please plot $v_\pi$ you obtained in (b) and (c). Add points $G_0$ to the same plot. Make sure to add title, label axes, and add legend indicating approximations used. Discuss. Which model you think performs better?

```python
q = td_2[0]
p = td_4[0]
```

```python
estimates = np.zeros((len(test),4))
T = 49
Î³ = .97

for k in range(len(test)):
    episode = get_episode(test, k)
    S_0 = episode[0][0]   
    G = 0
    for t in reversed(range(0,T)):
        R = episode[2][t]
        G = G*Î³ + R
    estimates[k,:] = S_0, G, q[S_0-1], p[S_0-1]
```

```python
estimates
```

```python
x = estimates[estimates[:,0].argsort()]
x.shape
```

```python
plt.figure(figsize=(16,6))
plt.scatter(range(x.shape[0]),x[:,1],color = "r", label='$G_0$')
plt.scatter(range(x.shape[0]),x[:,2],color = "b", label='$\hat{v}$ quadratic approx.')
plt.scatter(range(x.shape[0]),x[:,3],color = "g", label='$\hat{v}$ polynomial approx.')
plt.xlabel("k",fontsize=16)
plt.ylabel("$v_\pi(stage\,\,k)$",fontsize=16)
plt.title('Dependence of estimated $v_\pi(stage\,\,k)$ on $k$',fontsize=16)

#plt.xlim(0,18.5)
plt.grid(True)
plt.legend(loc = 'upper right',fontsize=16)
plt.show()
```

As we had seen in the "Dependence of estimated $v_\pi(stage\,\,k)$ on $k$" plot, the quadratic approximation is not a good fit for states > 15. The polynomial does a much better job of fitting the data, this can be expected as we would expect a better fit with each successive polynomial term, however normally this introduces the possibility of overfitting the data. In this case though, we can see that the quadratic is overestimating at the tail and the polynomial is a better fit overall, so we would expect it to have the lowest MSE.

### (e) Mean squared errors

(e) Please compute the mean squared error as follows

$$MSE = \frac{1}{1600}\sum_{i=1}^{1600} \left(G_0^i-\hat{v}(S_0^i,{\bf w})\right)^2$$

for both approximations (b) and (c). Here, $i$ indicates patients in the test set. Which model has smaller MSE? Discuss.

```python
def mse(target=estimates[:,1], val=estimates[:,2]): 
    return np.mean((target-val)**2)
    
print("MSE for quadratic approximation:", mse(val=estimates[:,2]))
print("MSE for polynomial approximation:", mse(val=estimates[:,3]))
```

Again we could see visually from the previous graph that the polynomial was a better fit for the $G_0$ especially at the tail. A quadratic function cannot capture the necessary curvature here.
