## Problem 1 (15 points)

Please consider a $4$-armed Bandit Problem with $q_*(1)=4$, $q_*(2)=6$, $q_*(3)=8$, $q_*(4)=10$, and a unit variance of rewards for each action. Using classes Environment and Agent, implement the Gradient Bandit Algorithm with $\alpha$ = 0.4. Initialize the numerical preference at 0 for all actions. Run the algorithm 500 times (i.e. for 500 episods) for T = 1000 and compute the observed average reward for each time $t$. Plot the average reward and the proportion of times the optimal action was selected for $t=1, 2,\ldots, T$.

```python
import random
from matplotlib import pyplot as plt 
import numpy as np
%matplotlib inline

class Environment:
    def __init__(self):
        self.time = 1
        self._means = (4, 6, 8, 10)
        
    def admissible_actions(self):
        return list(range(len(self._means)))

    def get_reward(self, action):
        self.time += 1
        return random.gauss(self._means[action],1)

class Agent:
    def __init__(self):
        self.H = np.zeros(4)
        self.r = 0.0
        self.r_bar = 0.0
        self.a = None
        
    def softmax(self,x):
        return np.exp(x - x.max()) / np.sum(np.exp(x - x.max()), axis=0)
    
    def step(self, env, alpha):
        pi = self.softmax(self.H)
        actions = env.admissible_actions()
        self.a = np.random.choice(actions, p=pi)
        self.r = env.get_reward(self.a)
        
        #Update average reward for action
        self.r_bar += (self.r - self.r_bar) / env.time
        
        # Update preferences
        self.H[self.a] += alpha * (self.r - self.r_bar) * (1 - pi[self.a])
        
        # For A!=a
        not_a = actions != self.a
        self.H[not_a] -= alpha * (self.r - self.r_bar) * pi[not_a]
        
        
def rewards_hist(T = 100, alpha=.4):
    env = Environment()
    agent = Agent()
    rewards = []
    while env.time <= T:
        agent.step(env, alpha)
        rewards.append(agent.r)
    return np.asarray(rewards)        
```

```python
T = 1000
episodes = 500
all_rewards = np.array([rewards_hist(T) for _ in range(episodes)])
avg_rewards = np.mean(all_rewards,axis=0)


plt.figure(figsize=(16,6))
plt.plot(range(1,T+1),np.transpose(all_rewards), linestyle='--', marker=None, markersize=4)
plt.plot(range(1,T+1),avg_rewards, linestyle='--', marker='o', color="r", markersize=4, label='Gradient Bandit ( = 0.4)')
plt.xlabel("t")
plt.ylabel("$R_t$")
plt.title('Dependence of average historical rewards $R_t$ on $t$ (T=1000) over 500 episodes')
plt.grid(True)
plt.legend(loc = 'upper right')
plt.show()
```

## Problem 2 (5 points)

Add "state" attribute, $S_t$, to the Environment which takes values 1 and 2. Assume that the agent still can take actions 1 through 4 in both states. If $S_t=1$, the agent gets rewards as in Problem 1, but if $S_t=2$ the Environment generates rewards with means -10, -8, -6, and -4 for actions 1, 2, 3, and 4, respectively (assume unit variance as before). Further, assume the Environment can change its state only if the agent selects action 1, and the probability of the switch is 0.05, that is:

$P(S_t = 2|S_{t-1} = 1, A_{t-1} = 1)=0.05$,   
$P(S_t = 1|S_{t-1} = 1, A_{t-1} = 1)=0.95$,   
$P(S_t = 1|S_{t-1} = 2, A_{t-1} = 1)=0.05$,   
$P(S_t = 2|S_{t-1} = 2, A_{t-1} = 1)=0.95$,

$P(S_t = 1|S_{t-1} = 1, A_{t-1} = a)=1$ for all $a\in\{2,3,4\}$, and     
$P(S_t = 2|S_{t-1} = 2, A_{t-1} = a)=1$ for all $a\in\{2,3,4\}$,

i.e. all other transition probabilities are 0.      

Please modify the Gradient Bandit Algorithm you developed in Problem 1 so that numerical preference and policy are state-specific now, i.e. you need to keep track of observations in each state separately. Also, please keep track of average rewards, the baselines, for each state separately. 

Plot results and discuss your observations. Does the Gradient Bandit Algorithm appear to be appropriate in the case of Markov Decision Process (MDP)? Explain.   
Hint: You may want to try running the algorithm once only (i.e. for one episode) in order to better understand what is going on.

```python
class Environment:
    def __init__(self, state=random.randint(0,1)):
        self.state = state
        self.time = 1
        self._means = [(4, 6, 8, 10),(-10, -8, -6, -4)]
        
    def admissible_actions(self):
        return list(range((4)))

    def get_reward(self, action):
        self.time += 1
        return random.gauss(self._means[self.state][action],1)
    
    def switch_state(self, action):
        # Switch states with a prob of .05
        if random.random() <= .05:
            self.state = 1 - self.state  
        

class Agent:
    def __init__(self):
        self.H = np.zeros((2, 4))
        self.r = 0.0
        self.r_bar = np.zeros(2)
        self.a = None
        
    def softmax(self,x):
        return np.exp(x - x.max()) / np.sum(np.exp(x - x.max()), axis=0)
    
    def step(self, env, alpha):
        pi = self.softmax(self.H[env.state])
        actions = env.admissible_actions()
        self.a = np.random.choice(actions, p=pi)
        self.r = env.get_reward(self.a)
        
        #Update average reward for action
        self.r_bar[env.state] += (self.r - self.r_bar[env.state]) / env.time
        
        # Update preferences
        self.H[env.state][self.a] += alpha * (self.r - self.r_bar[env.state]) * (1 - pi[self.a])
        
        # For A!=a
        not_a = actions != self.a
        self.H[env.state][not_a] -= alpha * (self.r - self.r_bar[env.state]) * pi[not_a]
        
        # Update state
        env.switch_state(self.a)
        
        
def rewards_hist(T = 100, alpha=.4):
    env = Environment()
    agent = Agent()
    rewards = []
    while env.time <= T:
        agent.step(env, alpha)
        rewards.append(agent.r)
    return np.asarray(rewards)        
```

```python
T = 1000
episodes = 500
all_rewards = np.array([rewards_hist(T) for _ in range(episodes)])
avg_rewards = np.mean(all_rewards,axis=0)


plt.figure(figsize=(16,6))
#plt.plot(range(1,T+1),np.transpose(all_rewards), linestyle='--', marker=None, markersize=4)
plt.plot(range(1,T+1),avg_rewards, linestyle='--', marker='o', color="r", markersize=4, label='Gradient Bandit ( = 0.4)')
plt.xlabel("t")
plt.ylabel("$R_t$")
plt.title('Dependence of average historical rewards $R_t$ on $t$ for 2 states (T=1000) over 500 episodes')
plt.grid(True)
plt.legend(loc = 'upper right')
plt.show()
```

We can see that with policy and transition matrix, we are not maximising the expected value of our reward. In 5% of our runs, we are switching to a state where expected reward is the opposite. Consequently we should expect that our reward approaches 3 ((10-4)/2) as this is the average of the optimal actions in each state. From a markov decision process, our algorithm seems appropriate as the current state is soley dependent on the previous state and action.
