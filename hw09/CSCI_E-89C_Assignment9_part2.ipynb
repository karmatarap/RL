In this problem we again consider patients with end-stage liver disease (ESLD). We assume that patient's health condition is fully characterized by the Model for End-stage Liver Disease (MELD) score (Jae-Hyeon Ahn and John Hornberger, Involving patients in the cadaveric kidney transplant allocation process: a decision-theoretic perspective. Manage Sci. 1996;42(5):629–41).

The MELD score ranges from 6 to 40 and is derived based on the probability of survival at 3 months for patients with ESLD. Data in ESLD is usually sparse and often aggregated into Stages. We assume that there are 18 stages based on the ESLD: Stage 1, Stage2, ..., Stage 18. The time step is 1 year and the actions in Stages 1 through 18 are "wait" (denoted by 0) and "transplant" (denoted by 1). 

We assume that the Markov property holds. There are two additional states of the Markov Decision Process: "Posttransplant Life" (denoted by 19) and "Death" (which is denoted by 20 and combines so caled "Pretransplant Death" and "Posttransplant Death"). The only action availible in state "Posttransplant Life" is "wait" and "Death" is the terminal state with no actions. Assume that the length of an episode is T=50, unless it terminates earlier due to the transition to the absorbing state "Death."

We do not know the transition probabilities, but if a patient selects "wait," the possible transitions are   
1) Stage 1->Stage 1, Stage 1->Stage 2, Stage 1->Death  
2) For k in {2,3,4,...17}, Stage k->Stage (k-1), Stage k->Stage k, Stage k->Stage (k+1), Stage k->Death    
3) Stage 18->Stage 17, Stage 18->Stage 18, Stage 18->Death    

If a patient selects "transplant" at Stage k, k=1,2,...,18, the only possible transition is  
4) Stage k->"Posttransplant Life"

Finally, there are two more possible transitions"  
5) "Posttransplant Life"->"Posttransplant Life" and "Posttransplant Life"->"Death"  


The patient gets reward 1 in all states "Stage k" (k=1,2,...,18) and reward 0.2 in the "Posttransplant Life" state - assume that the patient gets these rewards on "exit" from the states, i.e. after we observe the corresponding stage. We assume the discounting parameter $\gamma=0.97$, one of the most common discounting rate used in medical decision making (Gold MR, Siegel JE, Russell LB, Weinstein MC. Cost-Effectiveness in Health and Medicine. Oxford University Press; New York: 1996).


Please consider statistics on 8,000 patients with ESLD saved in the 'ESLD_statistics.csv' file. Eeach row represents an episode (i.e. one patient) and the columns are the sequences of the patients' states and actions. This data were generated under the behavor policy:

which means that, for example, 5% of paients at stage 14 received a transplant.

```python
from matplotlib import pyplot as plt 
import numpy as np
import pandas as pd
import math
import random
import time
```

```python
df = pd.read_csv('../ESLD_statistics.csv')
df.head(5)
```

```python
def get_episode(df, k):
    states = df.values[k,0:-1:2]
    actions = df.values[k,1::2]
    rewards = (states <= 18)*1 + (states == 19)*0.2
    return [states, actions, rewards]
```

```python
def argmax(q_values, admissable=[0,1]):
    """argmax with random tie-breaking (np.argmax always takes the first)
       Also allow for ignoring an index
    Args:
        q_values (Numpy array): the array of action-values
    Returns:
        action (int): an action with the highest value
    """
    top = float("-inf")
    ties = []

    for i in range(len(q_values)):
        if i in admissable:
            if q_values[i] > top:
                top = q_values[i]
                ties = []

            if q_values[i] == top:
                ties.append(i)

    return np.random.choice(ties)
```


$$q_\pi(s,a)\approx\hat{q}(s,a,{\bf w}) = \sum_{k=1}^{18} \left(\big[w_1+w_2\,a\big] + \big[w_3+w_4\,a\big]\frac{k-9}{5}\right)\cdot \mathbb{1}_{(s=\text{"Stage }k\text{"})}+w_5\cdot \mathbb{1}_{(s=\text{"Posttransplant Life"})},$$

where ${\bf w}=(w_1,w_2,w_3,w_4,w_5)^T$. 

```python
def q_hat(s, a, w):
    k = (s-9)/5
    return ((w[0] + w[1]*a) + (w[2]*k+w[3]*a*k))*(s<=18) + w[4]*(s==19)
```

```python
w = np.ones(5)
for s in range(1,21):
    for a in range(2):
        print("State: ", s, "Action:", a, q_hat(s,a,w))
```

```python
def dq_hat(s, a, w):
    #print(s,a,w)
    k = (s-9)/5
    dq = [1, a, k, a*k,0]*int(s<=18) + [0, 0, 0, 0, 1]*int(s==19) + [0, 0, 0, 0, 0]*int(s==20)
    return np.array(dq)


for s in range(1,21):
    for a in range(2):
        print("State: ", s, "Action:", a, dq_hat(s,a,w))
```

```python
def q_approx(α = "0.1", γ = 0.97, df=df, qa_hat=q_hat, dqa_hat=dq_hat, w_dim=5, w_init=None, num_epochs = 500000, seed=1234):
    """
    Q learning approximation algorithm with eligibility traces
    
    Args:
        α (char)         : Alpha computation that is evaluated, allows for fixed or flexible weight decays
        γ (float)        : discount factor between 0 and 1
        df (dataframe)   : pandas dataframe containing episodes
        qa_hat (function) : function to compute q_hat(s, a, w)
        dqa_hat (function): function that computes derivative of qa_hat
        w_dim (int or tuple)  : specify the dimension for w - 5 for gradient methods, (20,2) for tabular methods
        w_init (Bool or numpy array): initial weights
        num_epochs (int) : Number of bootstrap samples to take
        seed (int)       : random seed to use for reproducibility
        
    Returns:
        q_hat (numpy array) Estimates of V_hat
        w (numpy array) weights
        w_history (numpy array) History of Weights
    """
    np.random.seed(seed)
    N = 0
    w = np.zeros(w_dim)
    
    if not w_init is None:
        w_replay = w_init
    
    w_shape = (w_dim, 1) if type(w_dim) == int else w_dim
    
    w_history = w.reshape(w_shape)

    for _ in range(num_epochs):
        N += 1
        # sample the episode
        k = np.random.randint(0, df.shape[0])
        episode = get_episode(df, k)
        # sample the state
        t = np.random.randint(low=0, high=49)
        S = episode[0][t]
        S1 = episode[0][t+1]
        A = episode[1][t]
        R = episode[2][t]
        δ = (R + γ*max([qa_hat(S1, a, w) for a in range(2)])) - qa_hat(S, A, w)
        w = w + eval(α)* δ * dqa_hat(S, A, w)
        if (int(N/1000)*1000==N):
            w_history = np.append(w_history, w.reshape(w_shape), axis=1)
            
    q_vals = np.array([[qa_hat(k,a,w) for a in range(2)] for k in range(1,21)])
    return q_vals, w, w_history
```

```python
def plot_history(w_hist, α, title="Q learning (approx): Dependence of estimated $w_i$ on time step $t$ for $i=1,2,3,4$", ylabel="$w_\i$"):
    plt.figure(figsize=(16,6))
    
    for i in range(w_hist.shape[0]):
        plt.plot(range(1,w_hist.shape[1]*1000,1000),w_hist[i], linestyle='--', marker=None, markersize=4,label=f'$w_{i+1}$')
    plt.xlabel("time step $t$",fontsize=16)
    plt.ylabel(ylabel,fontsize=16)
    plt.title(f'{title} on time step $t$ for α={α}',fontsize=16)
    plt.grid(True)
    plt.legend(loc = 'upper left',fontsize=14)
```

```python
q_1 = q_approx(α="0.1")
```

```python
plot_history(q_1[2],.1)
```

```python
q_2 = q_approx(α="0.01")
plot_history(q_2[2],.01)
```

```python
q_3 = q_approx(α="0.001")
plot_history(q_3[2],.001)
```

```python
q_4 = q_approx(α="1/math.sqrt(N)")
plot_history(q_4[2],"1/sqrt(N)")
```

Based on the experimentation above, we see that 0.01 and $\frac{1}{\sqrt{N}}$  are good choices of $\alpha$. $\alpha$ values of 0.1 and 0.001 where too high and too low respectively. We could see that in the case where the step size was too large, there was no convergence. Alternatively, when the step size was too low, we saw convergence that was too slow for $w_2, w_4$, and $w_5$. The ´dynamic´ $\alpha$ value was good at initially taking larger steps and then taking progressively smaller steps to ensure a good separation. The *best* value was consequently $\frac{1}{\sqrt{N}}$, followed by 0.01.

```python
def plot_q_relative_overlay(qs=[],labels=(),colors=('r','g','b','y'),linestyles=(':','-','-.','--'),ylims=(0,1)):
    plt.figure(figsize=(16,6))
    for l in range(len(qs)):
        plt.plot(range(1,19),qs[l][0:18,0]/np.sum(qs[l][0:18,:],axis=1), linestyle=linestyles[l], marker="o",mfc='none',color = colors[l], markersize=8, label=f"Wait: {labels[l]}")
        plt.plot(range(1,19),qs[l][0:18,1]/np.sum(qs[l][0:18,:],axis=1), linestyle=linestyles[l], marker="o",mfc='none',color = colors[l], markersize=8, label=f"Transplant: {labels[l]}")
    plt.plot()
    plt.xlabel("k",fontsize=16)
    plt.ylabel(f"$q_π(stage\,\,k)$",fontsize=16)
    plt.title(f'Relative Dependence of estimated $q_π(stage\,\,k)$ on $k$',fontsize=16)
    plt.ylim(ylims)
    plt.xlim(0.5,18.5)
    plt.grid(True)
    plt.legend(loc = 'upper left',fontsize=12)
    plt.show()
```

```python
def tab_q(α = "1/math.sqrt(N)", γ = 0.97, df=df, num_epochs = 500000, seed=1234):
    """
    Tabular Q-learning
    
    Args:
        α (char)         : Alpha computation that is evaluated, allows for fixed or flexible weight decays
        γ (float)        : discount factor between 0 and 1
        df (dataframe)   : pandas dataframe containing episodes
        num_epochs (int) : Number of bootstrap samples to take
        seed (int)       : random seed to use for reproducibility
        
    Returns:
        Q (numpy array) Estimates of state action values
    """
    np.random.seed(seed)
    N = 0
    Q = np.zeros((20,2))
    for _ in range(num_epochs):
        N += 1
        # sample the episode
        k = np.random.randint(0, df.shape[0])
        episode = get_episode(df, k)
        # sample the state
        t = np.random.randint(low=0, high=49)
        S = episode[0][t]
        S1 = episode[0][t+1]
        A = episode[1][t]
        R = episode[2][t]
        Q[S-1,A] = Q[S-1,A] + eval(α)*(R + γ *max(Q[S1-1,:]) - Q[S-1,A])

    return Q

tab_q_vals = tab_q()
print(tab_q_vals)
```
