# Preface
We first consider images of 10 types (handwritten digits from "0" to "9") and build a feedforward fully connected Deep Neural Network for classification of the images (the accuracy, in fact, can further be increased by adding convolutional layers).  

Let's load libraries:

```python
from keras.datasets import mnist
from matplotlib import pyplot as plt 
import numpy as np
import pandas as pd
import time
import pickle

from keras import layers
from keras import models
from keras.layers import Dropout
import random
```

Next, we load the images:

```python
with open('handwritten_digits.pkl', 'rb') as f:
    X_train, y_train, X_test, y_test = pickle.load(f)
```

Here, 'X_train' and 'X_test' are the sets of 28x28 images; and 'y_train' and 'y_test' are the corresponding correct labels. For example, let's look at the first 18 images in 'X_train' dataset:

```python
fig = plt.figure(figsize=(15,8))
for i in range(18):
    plt.subplot(3,6,i+1)
    plt.tight_layout()
    plt.imshow(X_train[i], cmap='gray', interpolation='none')
    plt.title("Label: {}".format(y_train[i]),fontsize=16)
    plt.xticks([])
    plt.yticks([])
plt.show()
```

Let's reshape the images (we would not need to reshape in case of Convolutional Neural Network) which will be inputs to the Neural Network: 

```python
X_train = X_train.reshape(60000, 784)
X_test = X_test.reshape(10000, 784)
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

Let's also rescale the inputs:

```python
X_train /= 255
X_test /= 255
```

Because we want to classify, let's create dummies for the output:

```python
y_train = pd.get_dummies(y_train)
y_test = pd.get_dummies(y_test)
```

```python
y_train.head(3)
```

Let's now build a Neural Network and train it:

```python
model0 = models.Sequential()
model0.add(layers.Dense(512, activation='relu', input_shape=(784,)))                          
model0.add(Dropout(0.2))
model0.add(layers.Dense(512, activation='relu'))
model0.add(Dropout(0.2))
model0.add(layers.Dense(10, activation='softmax'))
          
model0.summary()
```

```python
nepochs = 35
model0.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')

history0 = model0.fit(X_train, y_train,
          batch_size=128, epochs=nepochs,
          verbose=1,
          validation_data=(X_test, y_test))
```

```python
plt.clf()
train_acc = history0.history['accuracy']
test_acc = history0.history['val_accuracy']

plt.figure(figsize=(16,6))
plt.plot(range(1,nepochs+1), train_acc, 'b', label='train accuracy',color="blue")
plt.plot(range(1,nepochs+1), test_acc, 'b', label='test accuracy',color="red")

plt.title('max(test accuracy) = ' + str(round(max(test_acc),3)),fontsize=16)
plt.xlabel('Epochs',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.grid(True)
plt.legend(loc = 'lower right',fontsize=16)

plt.show()
```

Let's now classify the images the Neural Network has not seen:

```python
y_test_hat = model0.predict_classes(X_test)
```

```python
fig = plt.figure(figsize=(15,8))
for i in range(18):
    plt.subplot(3,6,i+1)
    plt.tight_layout()
    plt.imshow(X_test.reshape((-1,28,28))[i], cmap='gray', interpolation='none')
    plt.title("Predicted label: {}".format(y_test_hat[i]),fontsize=16)
    plt.xticks([])
    plt.yticks([])
plt.show()
```

# Problem 1 (20 points)

In this problem we will develop a Q-learning algorithm for selecting the best treatment based on "medical" images. Each 128x128 image is a vector of 784 features: ${\bf x}=(x_1,x_2,\ldots,x_{784})^T$.

Please load the observations of features ${\bf x}(S_t)$, $t=0,1,\ldots$ ('images' variable), actions $A_t$, $t=0,1,\ldots$ ('actions' variable), and rewards $R_t$, $t=1,2,\ldots$ from the 'markov_decision_process.pkl' file. Please notice that in each state $s\in \mathcal{S}$, the set of admissible actions is $\mathcal{A}(s)=\{0,1,2\}$. Here, "0" states for no treatment and "1" and "2" are two treatments of different types. We assume that this is a Markov Decision Process, that is, the joint distribution of the next state $S_{t+1}$ and reward $R_{t+1}$ depens on the pair of current state and action $(S_t, A_t)$ only.

Please develop a Q-learning algorithm with approximation via a Deep Neural Network (NN). Please use the vector of fetures ${\bf x}(S_t)$ as an input to the NN. The output of the NN should be a vector of action-values $(\hat{q}(S_t,0,{\bf w}),\hat{q}\left(S_t,1,{\bf w}),\hat{q}(S_t,2,{\bf w})\right)^T$. Please feel free to use either feedforward fully connected NN or Convolutional NN.

Please 
- train the NN in mini-batches of size 128, that is, the weights ${\bf w}$ will be updated every 128 time steps
- update the target $R_{t+1}+\gamma \max_{ a}\hat{q}(S_{t+1},a,{\bf w})$ every 1280 time steps only, that is, not the most recent weights should be used to evaluate this taget  - this usually guarantees a better stability of the algorithm

Once the training is finished, please pick 18 images at random and display them together with optimal actions in the titles.

## Convert actions to dummies

```python
def to_dummy(l: list) -> np.ndarray:
    """Convert categorical list to dummy encoded numpy array"""
    arr = np.array(l)
    return np.eye(3)[arr]

assert np.array_equal(to_dummy(1),np.array([0., 1., 0.]))
assert np.array_equal(to_dummy(np.array([1, 2])) , np.array([[0,1,0], [0,0,1]]))
```

```python
actions_ = to_dummy(actions)
```

## Base Model

This is our simple base model that consists of 3 hidden layers, each with a relu activation and dropout of 0.2. Our output layer is of size one as we will expect a single estimate of q_hat.

```python
def define_model():
    model = models.Sequential()
    model.add(layers.Dense(512, activation="relu", input_shape=(787,)))                          
    model.add(Dropout(0.2))
    model.add(layers.Dense(512, activation="relu"))
    model.add(Dropout(0.2))
    model.add(layers.Dense(512, activation="relu"))
    model.add(Dropout(0.2))
    model.add(layers.Dense(1))

    model.summary()
    model.compile(loss='mse', metrics=['mse'], optimizer='adam')
    return model
```

For continuous estimates, the loss function we will use is "mean squared error"

```python
model = define_model()
```

```python
np_actions = np.array(actions)
X_zeros = xs_a[np_actions == 0,]
X_ones = xs_a[np_actions == 1,]
X_twos = xs_a[np_actions == 2,]
```

```python
X0 = np.concatenate((Xs, to_dummy([0]*Xs.shape[0])),axis=1)
X1 = np.concatenate((Xs, to_dummy([1]*Xs.shape[0])),axis=1)
X2 = np.concatenate((Xs, to_dummy([2]*Xs.shape[0])),axis=1)
```

```python
=.97
n_epochs=10
batch_size = 1280
mini_batch_size = 128
batches = len(images) // batch_size
for k in range(batches):
    times = np.arange(batch_size*k, batch_size*(k+1))
    q_hat_0 = model.predict(X0[times+1])
    q_hat_1 = model.predict(X1[times+1])
    q_hat_2 = model.predict(X2[times+1])
    q_hat = np.min(np.concatenate((q_hat_0, q_hat_1, q_hat_2),axis=1),axis=1)
    target = -np.array(rewards)[times] +  * q_hat
    
    history = model.fit(xs_a[times], target, batch_size=mini_batch_size, epochs=n_epochs, verbose=1)
```

```python
import numpy as np
from pytest import approx
W = np.array([np.array([[-1.2, .1, .5],[.9, .8, -.3]]), np.array([.2, -.8, 1.2])])
```

```python
x = np.array([1, 1.3, .7])
```

```python
relu = lambda x: np.maximum(x,0) #relu
relu_dv = lambda x: (x>=0) + 0 #partial derivative of relu
```

```python
def forward_and_back(W, x, v_pi=4.1, alpha=.1):
    #hidden layer 1
    z1 = np.dot(W[0][0], x)
    u1 = relu(z1)
    assert u1 == 0.
    
    #hidden layer 2
    z2 = np.dot(W[0][1], x)
    u2 = relu(z2)
    assert u2 == 1.73
    
    #output layer
    u = np.array([1, u1, u2])
    z = np.dot(W[1], u)
    v_hat = relu(z)
    assert round(v_hat,3) == 2.276
    
    #errors output layer
    e = v_hat - v_pi
    assert round(e, 3) == -1.824
    
    #errors hidden layer 1
    e1 = np.dot(e,np.dot(relu_dv(z2),W[1][1]))
    assert round(e1, 4) == 1.4592
    
    #errors hidden layer 2
    e2 = np.dot(e,np.dot(relu_dv(z2),W[1][2]))
    assert round(e2, 4) == -2.1888
    
    #partial derv output layer
    dv = np.dot(e, relu_dv(z) * u)
    assert dv == approx(np.array([-1.824, 0., -3.15552]))
    
    #derivative for hidden layer 1
    dv1 = np.dot(e1, np.dot(relu_dv(z1), x))
    assert dv1 == approx(np.array([0.,0.,0.]))
    
    dv2 = np.dot(e2, np.dot(relu_dv(z2), x))
    assert dv2 == approx(np.array([-2.1888,-2.84544,-1.53216]))
    
    #sgd update
    w = np.concatenate((W[0][0], W[0][1], W[1]))
    deltaL = np.concatenate((dv1, dv2, dv))
    update = w - alpha * deltaL
    assert update == approx(np.array([-1.2, 0.1, 0.5, 1.11888, 1.084544, -0.146784, 0.3824, -0.8, 1.515552]))
    
    
forward_and_back(W,x)    
```
